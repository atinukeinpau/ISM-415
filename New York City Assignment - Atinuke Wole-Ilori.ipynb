```python
import pandas as pd
```

```python
data = pd.read_csv('NYC_Property_Data.csv')
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.dropna(how='any', inplace=True)
```

```python
features = data.drop(['Unnamed: 0', 'Unnamed: 0.1', 'ADDRESS', 'SALE PRICE', 'SALE DATE'], axis=1)
```

```python
#features['SALE DATE'] = pd.to_datetime(features['SALE DATE'], errors = 'coerce')
```

```python
features
```

```python
target = data['SALE PRICE']
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
features['NEIGHBORHOOD'] = le.fit_transform(features['NEIGHBORHOOD'])
```

```python
features['BUILDING CLASS CATEGORY'] = le.fit_transform(features['BUILDING CLASS CATEGORY'])
```

```python
features['TAX CLASS AT PRESENT'] = le.fit_transform(features['TAX CLASS AT PRESENT'])
```

```python
features['BUILDING CLASS AT PRESENT'] = le.fit_transform(features['BUILDING CLASS AT PRESENT'])
```

```python
features['BUILDING CLASS AT TIME OF SALE'] = le.fit_transform(features['BUILDING CLASS AT TIME OF SALE'])
```

```python
features
```

```python
from sklearn.model_selection import train_test_split
```

```python
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size = 0.2)
```

```python
features_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
model = regressor.fit(features_train, target_train)
```

```python
target_pred = model.predict(features_test)
```

```python
pd.DataFrame({'Actual': target_test, 'Predicted': target_pred, 'Error': target_test - target_pred})
```
